```python
def generate_mean_trajectories(normal_distribution, number_of_trajectories, length_of_trajectory):
    def genList():
        a = [normal_distribution.gen_rand() for i in range(length_of_trajectory)]
        sum = 0
        cumsum = [sum := (sum + s)/(a.index(s)+1) for s in a]
        return cumsum
    return [genList() for i in range(number_of_trajectories)]
```

```python
generate_mean_trajectories(d.NormalDistribution(random, 0, 1), 3, 5)
```

```python
def linePlot(input_list):
    x = [i for i in range(len(input_list))]
    y = input_list
    for i in range(len(y[0])):
        plt.plot(x,[pt[i] for pt in y],label = 'id %s'%i)
    plt.legend()
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("A test graph")
    plt.show()
```

```python
linePlot(generate_mean_trajectories(d.NormalDistribution(random, 0, 1), 3, 5)
)
```
